```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
```

```python
#Task 1 
#Sort the data as per total number of cases
sorted_df = df.sort_values(by=("total_cases"))
sorted_df
```

```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph

total_cases = sorted_df['total_cases'].head(5)
total_cases

country = sorted_df['country'].head(5)
country

label = total_cases
value = country
plt.xticks(rotation='vertical')
plt.bar(value, label, width=0.4, color=('red','yellow','green','pink','blue'))

total_cases
country
```

```python
#Task 3
#Sort the data as per total number of deaths

total_death = df.sort_values(by=("total_deaths"))
total_death
```

```python
#Task 4
#Get top 5 countries who has the maximum number of deaths and plot a bar graph
death = total_death['total_deaths'].tail(5)
dcountry  =total_death['country'].tail(5)

plt.xticks(rotation='vertical')

plt.bar(dcountry, death, width=0.4, color=('red','yellow','green','pink','blue'))
```

```python
#Task 5
#Sort the data as per active cases

active_case = df.sort_values(by=("active_cases"))
active_case
```

```python
#Task 6
#Get top 5 countries who has the maximum number of active cases and plot a bar graph
case = active_case['active_cases'].tail(5)
acountry  =active_case['country'].tail(5)

plt.xticks(rotation='vertical')

plt.bar(acountry, case, width=0.4, color=('red','yellow','green','pink','blue'))


print(acountry)
case
```
